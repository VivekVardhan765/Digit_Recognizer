```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_data=pd.read_csv("train.csv")
test_data=pd.read_csv("test.csv")
```

```python
test_data.head()
```

```python
train_data.head()
```

```python
train_target=train_data.iloc[:,0]
```

```python
train_data=train_data.iloc[:,1:]
train_data.shape
```

```python
import matplotlib.pyplot as plt
plt.imshow(train_data.iloc[1234].values.reshape(28,28))
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(train_data,train_target,test_size=0.2,random_state=42)
```

```python
X_train.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
knn3 = KNeighborsClassifier(n_neighbors=3)
knn3.fit(X_train,y_train)
knn5 = KNeighborsClassifier(n_neighbors=5)
knn5.fit(X_train,y_train)
knn7 = KNeighborsClassifier(n_neighbors=7)
knn7.fit(X_train,y_train)
```

```python

y_pred = knn3.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
y_pred = knn5.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
y_pred = knn7.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
r_test=test_data.sample(10)
r_test.shape
predicted_output=knn3.predict(r_test)
for t in predicted_output:
    plt.imshow(r_test.iloc[t].values.reshape(28,28))
    plt.show()
    print("Predicted Digit =", str(predicted_output[t]), "\n")
```
